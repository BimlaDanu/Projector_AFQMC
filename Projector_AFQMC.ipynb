### Projector AFQMC Algorithms


 - A second-order symmetric Trotter-Suzuki decomposition is used in our Monte-Carlo sampling approach for  imaginary time propagators

    - $e^{-\Delta_\tau \hat{H}_\text{AFQMC}} = e^{-\frac{\Delta_\tau}{2} \hat{H}_t} e^{-\Delta_\tau \hat{H}_\text{int}}   e^{-\frac{\Delta_\tau}{2} \hat{H}_t}  + \mathcal{O}(\Delta_\tau^3)$
        - Here,  $\Delta_\tau = \beta/L_\tau$ with  $L_\tau$ the number of Trotter slices
        -  $\beta=\frac{1}{ k_B T}$ is the inverse temperature

 -  In the zero temperature projector algorithm, the ground state  $|\psi_0\rangle$ is obtained from the imaginary time projection  $|\psi_0\rangle =  \lim\limits_{\Theta\to\infty} e^{-\Theta {\hat H}}  |\psi_T\rangle \equiv  \lim\limits_{L_\tau\to\infty}
     (e^{-\Delta\tau \hat H})^{L_\tau} |\psi_T\rangle$
    - $|\psi_T\rangle$ is typically a Slater-determinant trial state with $\langle \psi_T|\psi_0 \rangle\ne0$  
    - $2\Theta =  L_\tau \Delta \tau$ is a projection parameter with  number Trotter slices $L_\tau$ and  Trotter steps $ \Delta \tau$

 - Monte Carlo(MC) sampling:
$
\langle \hat H\rangle  = \frac{\big\langle \psi_0 \big| {\hat H} \big| \psi_0 \big\rangle}{\big\langle \psi_0 \big| \psi_0 \big\rangle}  \equiv\frac{ \sum_C |W(C)| ~~\text{sign}(C) E_{\text{loc}}(C) }  { \sum_C |W(C)| \text{sign}(C)}$ where  C  corresponds to the Monte-Carlo configurations 


    - $\text{sign}(C) = \frac {W(C)} {|W(C)|}$

    - MC weight  $W(C) = \langle \psi_T^{\boldsymbol\alpha}| U^{\boldsymbol\alpha}_C(2\beta,0) | \psi_T^{\boldsymbol\alpha}\rangle $ 
    
    - Imaginary time propagators $U_C(\beta,0) = \prod_{\tau=1}^{L_\tau}
     e^{-\Delta\tau\, \hat{\boldsymbol c}^\dagger \hat H_{\mathrm{int}} \hat{\boldsymbol c}}
     e^{-\Delta\tau\, \hat{\boldsymbol c}^\dagger \hat H_t \hat{\boldsymbol c}}$ 
     
     - Local energy reads; $E_{\mathrm{loc}}(C) = \langle \psi_T^{\boldsymbol\alpha}| U^{\boldsymbol\alpha}_C(2\beta,\beta) {\hat H} U^{\boldsymbol\alpha}_C(\beta,0)| \psi_T^{\boldsymbol\alpha}\rangle/W(C)$ 


 -   General observables in the projective algorithm  are evaluated as, 
 $ \big\langle  {\hat O} \big\rangle  =\lim\limits_{\Theta \rightarrow \infty} \frac{\big\langle \psi_T \big| e^{-\Theta {\hat H}}  {\hat O}  e^{-\Theta {\hat H}}\big |\psi_T \big\rangle}{\big\langle \psi_T \big|e^{-2\Theta {\hat H}}\big|\psi_T \big\rangle}$
